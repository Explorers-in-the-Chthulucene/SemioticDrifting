Topic modelling for further vocabulary definition

We adopted <a href="https://github.com/MaartenGr/BERTopic">BERTopic</a> to extract clusters of topics for each chapters.

BERTopic has been particularly useful since it allows keywords extraction as well as hierarchical clustering of topics. In our case, given the disseminated and non-structured knowledge of <i>Staying with the trouble</i> it has been particularly helpful in clustering together all the different terms Haraway uses to describe a a specific concept.

### tokenization

We started by cleaning the pdf-text from all the apex notes, images, decorations and extracting it as txt document, then we transformed it into lowercase using <a href="https://pypi.org/project/clean-text/">clean text</a> python libraries.

```python
sentence_model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = sentence_model.encode(docs, show_progress_bar=False)

# Train BERTopic
topic_model = BERTopic().fit(docs, embeddings)

# Run the visualization with the original embeddings
visualization = topic_model.visualize_documents(docs, embeddings=embeddings)
```
