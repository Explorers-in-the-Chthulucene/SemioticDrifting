```python
import rdflib
from rdflib import *
from rdflib.term import Literal
from rdflib.graph import Graph
from pprint import pprint
```

```python
chthuluGraph = Graph()
chthuluGraph.parse("/content/Chthulucene_5.ttl", format="ttl")
```

```python
namespacestring="http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene"
chthulu = Namespace(namespacestring)
```

```python
owl = Namespace("http://www.w3.org/2002/07/owl#")
rdf = Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
xml= Namespace("http://www.w3.org/XML/1998/namespace")
xsd = Namespace("http://www.w3.org/2001/XMLSchema#")
rdfs = Namespace("http://www.w3.org/2000/01/rdf-schema#")
```

```python
#add the prefix to the graph
chthuluGraph.bind("chthulu", chthulu)
chthuluGraph.bind("owl", owl)
chthuluGraph.bind("rdf", rdf)
chthuluGraph.bind("xml", xml)
chthuluGraph.bind("xsd", xsd)
chthuluGraph.bind("rdfs", rdfs)
```

```python
# prova
print(chthuluGraph.serialize(format="ttl"))
```

```python
# 1 What are good practices to make kin?
cq1 = '''
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX chthu: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT ?subclass
WHERE {
  ?subclass rdfs:subClassOf chthu:Frame ;
           chthu:evokes ?concept .
  ?concept rdf:type chthu:Concept ;
           rdfs:label "making kin" .
}
'''
```

```python
# 6. What neologisms does Donna Haraway mint and what is their meaning?
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT DISTINCT ?concept
WHERE {
  ?frame rdf:type ontology:Frame ;
         ontology:hasRole [ ontology:agentiveRole ?agentRole ] .
  ?agentRole ontology:hasAgentiveEntity ?agentEntity .
  ?agentEntity rdf:type ontology:AgentiveEntity ;
               rdfs:label "Donna Haraway" .
  ?concept rdf:type ontology:Concept ;
           ontology:evokedBy ?frame .
}
```

```python
# 7. From what perspective is the story of x narrated?


PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT ?agentiveRole
WHERE {
  ?agentiveEntity rdf:type ontology:AgentiveEntity ;
                  ontology:hasAgentiveRole ?agentiveRole ;
                  ontology:involvesRole ontology:x .
}
```

```python
# 8. In what context did the sympoiesis between x, y, z? What was the consequence of that?


PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT ?conceptClass ?frame ?role ?agentiveEntity
WHERE {
  ?concept rdf:type ?conceptClass ;
           ontology:evokedByFrame ?frame .

  ?frame ontology:involvesRole ?role .

  ?agentiveEntity ontology:hasAgentiveRole ?role .
}
```

```python
# 9. What are different types of SF? What is SF?
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT DISTINCT ?property
WHERE {
  ?property rdf:type rdf:Property ;
            rdfs:domain ontology:PlayingSF .
}
```

```python
# 10. Who is the most quoted person in the book?

PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT ?agentiveEntity (COUNT(?agentiveEntity) AS ?count)
WHERE {
  ?agentiveEntity rdf:type ontology:AgentiveEntity .
}
GROUP BY ?agentiveEntity
ORDER BY DESC(?count)
LIMIT 1
```

```python
# 11. What are the fictional stories invoked to rethink reality?

PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT ?frameClass ?hasInstance ?hasLabel ?hasFramenetID ?subclassOf
WHERE {
  ?frameClass rdf:type rdfs:Class ;
              rdfs:subClassOf* ontology:Frame ;
              ontology:hasInstance ?hasInstance ;
              ontology:hasLabel ?hasLabel ;
              ontology:hasFramenetID ?hasFramenetID ;
              rdfs:subClassOf ?subclassOf .
}
```

```python
# 12. What is the book written by Anna Tsing that informed Haraway’s thinking?
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT ?descriptorClass
WHERE {
  ?humanProp rdf:type ontology:HumanProp ;
             rdf:label "Anna Tsing" .
  ?descriptorClass rdf:type ontology:Descriptor ;
                   ontology:descriptorRole ?humanProp .
}
```

```python
# 14. What is tentacular thinking?
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT ?instance ?property ?value
WHERE {
  ?instance rdf:type ?class .
  ?class rdfs:label "Tentacular Thinking" .
  ?instance ?property ?value .
}
```

```python
# 13. What are examples of Science-Art worldings for staying with the trouble?
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX ontology: <http://www.semanticweb.org/ghiot/ontologies/2023/5/chthulucene#>

SELECT DISTINCT ?frame
WHERE {
  ?instance rdf:type ontology:wordling .
  ?frame ontology:evokes ?instance .
}
```

```python
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint_url = "https://dbpedia.org/sparql/"

query = """
prefix dbr:  
SELECT ?city ?country
WHERE {VALUES ?city {dbr:Munich dbr:Graz} ?city dbo:country ?country}
"""

sparql = SPARQLWrapper(endpoint_url)
sparql.setQuery(query)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()

for result in results["results"]["bindings"]:
    print(result["country"]["value"])
```

```python
dbo = Namespace("http://dbpedia.org/ontology/")
chthuluGraph.bind("dbo", dbo)
for result in results["results"]["bindings"]:
  if "Germany" not in result["country"]["value"]:
    chthuluGraph.add((URIRef(result["city"]["value"]), dbo.Country, URIRef(result["country"]["value"])))
  else:
    chthuluGraph.add((URIRef(result["city"]["value"]), dbo.Country, URIRef(result["country"]["value"].replace("http://dbpedia.org/resource/", "", 1))))


print(chthuluGraph.serialize(format="ttl"))
```
